```python
#importing pandas

import pandas as pd
import numpy as np

```

```python
#Series from lists

company=['JPMorganChase & Co.','Fastenal','Visa','Cisco','Amazon','Uber','Oracle']

pd.Series(company) #series object contains {index,value}.
```

```python
#Numbers
marks=[96,98,99,99.5,100]

pd.Series(marks)
```

```python
#custom index
marks=[96,98,99,97,100]
subjects=['Chemistry','Physics','Mathematics','DSA','English']

pd.Series(marks,index=subjects)
```

```python
#setting a name
marks=pd.Series(marks,index=subjects,name='Marks of Shristy')
marks
```

```python
#Series from dictionary:-

marks={
    'Maths':98,
    'English':96,
    'Chemistry':97,
    'Physics':99,
    'Communication':100
}

marks_series=pd.Series(marks,name='Yash ke marks')
marks_series
```
