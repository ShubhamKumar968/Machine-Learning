```python
#(1). Sigmoid function:-  s(x)= 1/(1+e-x);
import numpy as np
def sigmoid(array): #sigmoid always gives result between (0,1)
    return 1/(1+np.exp(-(array)))#gives result of sigmoid of all elements of an array

a=np.arange(10)
sigmoid(a)
```

```python
#(2). Mean squared error:- mse(x)=1/n * sum(x-x')^2

actual=np.random.randint(1,50,25)
print(actual)

predicted=np.random.randint(1,50,25)
print(predicted)
```

```python
#mean squared error:-(loss function)
def MSE(actual, predicted):
    return np.mean((actual-predicted)**2)

MSE(actual,predicted)
```

```python
#(3).Binary cross entropy (log loss):-

#log Loss=−1/n∑​[yi​log(pi​)+(1−yi​)log(1−pi​)]
import numpy as np

def bce(y_true, y_pred):
    y_pred = np.clip(y_pred, 1e-15, 1 - 1e-15)# clip ensure p never becomes exactly 0 or 1avoids math errors
    return -np.mean(y_true * np.log(y_pred) + (1 - y_true) * np.log(1 - y_pred))

bce(actual,predicted)
```
