```python
#sets:(1)It is a mutable(changable) object
#(2)set can't have duplicate values
#(3)set do not have indexing/slicing
#(4)set do not allow mutable data type
#(5)set can store value in random order not as inserted (it follow hashing)

st={'hello',4.5,True,9}
print(type(st))

st1={1,1,2,3,4,5,4,3,2,2,1}
print(st1)

st2={(1,2,3,), 'hello'}
print(st2)

st3={[1,2,3],'hello'}
print(st3)

st4={{1},{3}}
print(st4)#can't create 2D set

```

```python
#edit or change set item

s={1,2,3,4}
s[0]=100
print(s)
```

```python
#add and remove item in set
s={1,2,3,4}
print(id(s))

s.add(100)
print(s)
print(id(s))#same id as above so it can be done(mutable)

s.remove(4)
print(s)

s.pop()#remove last element
print(s)
```

```python
#set operations

s1={1,2,3}
s2={4,5,'hello'}

#There is no (s1+s2) and (s1*5) possible for set

print(1 in s1)

for i in s1:
    print(i, end=" ")
    
```

```python
#set functions
s1={1,2,3,0,4}
s2={4,5,'hello'}
print(len(s1))

print(min(s1))

print(sum(s1))

print(sorted(s1))

print(s1.union(s2))

print(s1.difference(s2))

print(s1.intersection(s2))

print(s1.symmetric_difference(s2))

print(s1.isdisjoint(s2))

print(s1.issubset(s2))
```
