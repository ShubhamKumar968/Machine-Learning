```python
#literals are raw data given to a variables

#(1) Numeric literals

a=0b101 #binary literals
b=100   #decimal literals
c=0o310 #Octal literals
d=0x12c #Hexadecimal literals

print(a,b,c,d)

#float literals
float_1= 10.5
float_2=1.5e2
float_3=1.2e-3

print(float_1, float_2, float_3)

#complex literals
x=3+4j
print(x, x.imag, x.real)


#(2) string literals
a='This is python'
str="""This is used for multiline """
print(str)

unicode=u"\U0001f600"
print(unicode)


#(3) Boolean literals
a=True + 4
b=False + 10

print("a: ", a)#true=1; false=0
print("b: ", b)

#Special literals
a=None #to declare a variable safely, will no give error , if we do not need to assign variable.
print(a)
```
