```python
#Strings- It is sequence of Unicode(16 bits representation but ascii uses- 8 bit)  characters.

#(1) creating string

#String is immutable object can't be change or edited
s='Hello Googler'
print(s)

s1="I am an employee"
print(s1)

s2='''It's amazing!'''
print(s2)

s3= str("Hello")
print(s3)

s[0]=G
print(s)
```

```python
#(2) Accessing string

#(i) Indexing

s="Computer"
print(s)
print(s[3])#positive indexing(left to right {start from 0})
print(s[-1])#negative indexing(right to left {start from -1}
print(s[9])
```

```python
#(ii) Using for loop

s="I am back"
for i in s:
    print(i,end=" ")
```

```python
#(iii) Using Slicing operators :- object[st : end : step]
#By default step=1;

s="machine@Learning"

print(s[0 : 4])

print(s[2: ])

print(s[ :4])

print(s[1: :1])

print(s[ : 0 : -1])

print(s[5: : -1])

print(s[ : : -1]) #Reverse of string

print(s[ : : ])

print(s[-5 : -1: 2])

print(s[0 : 6 : -1])

print(sorted(s))

print(len(s))

print(max(s))

print(type(s))
print(s.index('i'))
print(s.count('i'))
print(s.startswith('M'))

print(s.upper())
print(s.replace('M', 'm'))
print(s.isdigit())

#Note:- String is immutable object so we can't edit or delete string.
```

```python
#(3)Operations on string

s1='Hello'
s2='World'

print(s1+s2) #concatenation
print(s1 * 5) #multiplication(To generate pattern)
print('*' * 50)


print('Hello'=='World')

print('Hello' != 'World')

print('Mumbai' > 'Pune') #lexicographically comparision

print('' and 'hii')
print('' or 'hey')
print('hello' and 'world') #empty string=false; non-empty=true
print('hello' or 'world')


c='maths'
for i in c[: : -1] :
    print(i, end=' ')

print(' ')
print('H' in c)
print('maths' not in c)

```

```python
#(4) Functions in strings

s='hello'

print(len(s))

print(max(s))#Based on ascii value

print(min(s))

print(sorted(s, reverse=True))#based on ascii value


s1='it is suMmer'
print(s1.capitalize())# only 1st letter ko capital banayega

print(s1.title())# each word ke 1st letter ko capital kaarega

print(s1.lower())

print(s1.upper())

print(s1.swapcase())#lower koo upper and upper ko lower me convert karega

print('it is raining'.count('rain'))

print('it is raining'.find('rain')) #find nhi milne par -1 return karta hai

print('it is raining'.index('rain')) #lekin index error show karta hai


print('It is raining'.endswith('it'))
print('It is raining'.startswith('It'))


```
