```python
#Operators are used to perform operations on the variables and value

#1. (Arithmetic Operators)

a=5
b=2

print(a+b) #addition
print(a-b) #subtraction
print(a*b) #product
print(a/b) #true division
print(a//2)#integer division
print(a%b) #remainder(modulus)
print(a**b)#poweof operator

print(3/4)
print(3.0/4)
print(3//4)
print(3.0//4)
print(3.5%2)  #modulo works on decimal also
print(-3%2)
```

```python
#2. (Relational or Comparision Operators)
a=5
b=2

print(a>b)
print(a<b)
print(a>=b)
print(a<=b)
print(a==b)
print(a!=b)
```

```python
#3. (Logical Operators)

x=True
y=False

print(x and y)
print(x or y)
print( not x)
```

```python
#4. (Bitwise Operators) -> Works on binary only
a=2 #(10)
b=3 #(11)

print(a & b) #and
print(a | b) #or
print(a^b) #xor
print(~a) #1's complement
print(a>>2)#right shift
print(b<<3)#left shift
```

```python
#5. (Assignment Operators)

a=3
print(id(a)) #gives address of variable
print(a)
a+=3
print(a)
a*=3
print(a)

#There is no ++ and -- operator in python
```

```python
#6. (Identity Operators) -> works only on identifier
a=3
b=3
print(a is b) #(yah batata hai ki do element same memory location par hai ya nhi)
x=[1,2,3]
y=[1,2,3]
print(x is not y)
```

```python
#7. (Membership Operators) -> works on containers {list,str,range,set,tuple}

x="Delhi"
print('D' not in x)

y=(1,2,3,)
print(5 in y)
```

```python
#Some examples

print(6>5>4)
print(not 3>4)
print('seeta' or 'geeta') #depends on 1st
print('Ram' and 'mohan') #depends on 2nd
print(3>0 and 4>0)
```
