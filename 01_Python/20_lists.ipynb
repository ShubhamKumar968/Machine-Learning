```python
#list:- It is a mutable(changable) object, and can store heterogeneous elements, No contiguous memory address, so it is slower than array

l1=['hello',4,True,6.7,5+6j] #1D
print(l1)

l2=[1,2,3,[4,5]]#2D
print(l2)

l3=[[[1,2],[3,4]], [[5,6], [7,8]]]#3D
print(l3)

l4=list('patna') #it takes only iterables as input.
print(l4)

l5=list(range(3))
print(l5)
```

```python
#Access list elements
l=[1,2,4,9,5,[2,9]]
l1=[[[1,2],[3,4]], [[5,6], [7,8]]]

print(l[3])
print(l[1], l[0])

print(l[5][1])

print(l[-1])#from reverse

print(l1[1][0][1])
```

```python
#Edit list (Because it is mutable)

l=[1,2,3,4,5]
l[2]=100
print(l)
```

```python
#Add element in list:
lis=[1,2,3,4,5]

lis.append(100)# append 1 item in last
print(lis)

lis.extend([10,80,90])# append multiple item in last
print(lis)

lis.extend('goa')
print(lis)

lis.insert(1,500)#append item at its desired location
print(lis)
```

```python
#Delete element in list:

l=[1,2,3,4,5,'goa']

del(l[0])#delete element at index
print(l)

l.remove(4)#delete if element is present
print(l)

l.pop()#delete last item
print(l)

l.clear()#empty the list
print(l)
```

```python
#Operations on lists

l1=[1,2,[3,10]]
l2=[9,7,8]

print(l1+l2)
print(l1==l2)
print(l1>l2)
print(l1*3)

print(l1)

for i in l1:
    print(i)

print(3 in l1)
```

```python
#BuiltIn methods of lists

l=[1,2,[3,10]]
l1=[5,1,2,3]
print(len(l))

print(max(l1))
print(min(l1))

l1.reverse()
print(l1)

print(l1.index(1))

print(sorted(l1)) #Not permanent operation(create new list in memory)
print(l1)

l1.sort() #  make change in original list
print(l1)

print(l1.count(3))

print(sum(l1))
```

```python
#unpacking

l1=[1,2,3]
a,b,c=l1
print(a,b,c)

#pacinkg
a=6; b=9;c=10
l2=[a,b,c]
print(l2)
```

```python
#List Compression

l= [a+1 for a in range(5)]
print(l)

l1=[i**2 for i in range(2,10,2)]
print(l1)
```
