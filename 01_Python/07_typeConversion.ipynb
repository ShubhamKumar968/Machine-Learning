```python

#TypeConversion -> it is used to convert one data type to another if it is possible

#(1)implicit typeConversion -> done automatically by python
4 + 5.5
4.5 + 5+ 5j
```

```python
#(2)Explicit TypeConversion -> done manually by using  builtIn function

a='5'
print(type(a))
print(int(a))
print(type(a))
print(bool('3'))
print(float(4))
print(str(9))
print(complex(6))
print(list("Hello"))
print(int('Patna'))

#type conversion is a temporary change -> i.e, data type is same as initial
```

```python
#program to add two numbers

first_num= int ( input("Enter the first number: ") )
second_num= int ( input("Enter the second number: ") )
print("first Num: ",first_num)
print("Second Num: ",second_num)
print("Sum is: " , (first_num + second_num)) #concatenation due to string type
print(type(first_num))
```
