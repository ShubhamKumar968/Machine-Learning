```python

#(1) Break:- it is use to terminate loop at a specific condition(linear search)

print("Visualizing break: ")
for i in range(1,11):
    if i==5:
        break
    print(i)


#(2) continue:- It is used to skip particular iteration only not all.
print("Visualizing continue: ")
for i in range(1,11):
    if i==5:
        continue
    print(i)

#(3)pass -> it is used when we want to run loop but currently we don't know logic
print("Visualizing pass: ")
for i in range(1,11):
    pass
```
