```python
                #Learning dataTypes
    
#Range of integer= 1e309 => {10^309 == infinity}

#(1) integer
print(4)
print(1e308)
print(1e309)
```

```python
#(2) float

print(4.5)
print(1.7e308)
print(1.7e309)
```

```python
#(3)  Boolean

print(True)
print(False)
```

```python
#(5) Complex

print(3+4j)
```

```python
#(6) String

print('Patna')
print("Patna")
print("""Patna""")

```

```python
#(2) tuple

print((1,2,3,4,5,))
```

```python
#(4) dictionary  -> same as map (her searching is very fast)

print({"Name" : 'Shubham' , "Age" : 20 , "Gender" : "Male" ,"Business":"Software Engineer"})
```
