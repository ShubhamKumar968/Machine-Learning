```python
#for loop is iterated on either sequence or range

#Agar hame pata hai ki loop kitna baar chalega to for loop , agar nhi pata to while loop use karenge.
print("Number and its square are:- ")
r=range(2,10,3)
for e in r:
    print(e, e**2)

print("Accessing range elements:- ")

for i in range(1,10):
    print(i)

print("Elements in string are:-")

for ch in "Placements" :
    print(ch)


#nested loop
rows=int(input("Enter number of rows:- "))

for i in range(1,rows+1):
    for j in range(0,i):
        print("*", end=' ')
    print(" ")
```
