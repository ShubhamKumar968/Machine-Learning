```python
#lambda functions:-

x=lambda x:x**2
print(x(9))
print(type(x))

#(1)lambda has no return value
#(2)it is written in one line
#(3)Not used for code readability
#(4)It has no any name.
#(5)It is used with higher order functions (function requres another function for its working)
```

```python
a=lambda x,y: x+y
print(a(4,5))
```

```python
b=lambda x:x[0]=='a'
print(b('apple'))
print(b('banana'))
```

```python
c=lambda x:'Even' if x%2==0 else 'Odd'
print(c(5))
```

```python
L=[11,14,21,23,56,78,45,29,28]
#higher order function(HOF)

x=lambda x:x%2==0
y=lambda y: y%2!=0
z=lambda z: z%3==0

def return_sum(func,L):
    result=0

    for i in L:
        if func(i):
            result = result+i
    return result

print(return_sum(x,L))
print(return_sum(y,L))
print(return_sum(z,L))
```

```python
#map function in hof

L=[1,2,3,4,5,7]

map(lambda x: x*2, L) #takes(lambda function, iterable)
```

```python
print(list(map(lambda x: x*2, L)))
print(list(map(lambda x: x%2==0, L)))
```

```python
#Filter:- (it works based on certain condition)

list(filter(lambda x: x>4,L))
```

```python
#Reduce:-

import functools #(for reduce)

L=[1,2,3,4,5,6,7]

print(functools.reduce(lambda x,y:x+y,L))
```

```python
#List comprehension
L=[1,2,3,4,5,6,7]

L1=[item*2 for item in L]
print(L1)

print([i**2 for i in range(10) if i%2!=0])
```

```python
#dictionary comprehension

d={'name':'Ram', 'Gender':'male', 'Age':20}
print(d.items())
```

```python
d1={key:value for key,value in d.items() if len(key)>3}
print(d1)
```
