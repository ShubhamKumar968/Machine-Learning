```python
#create an Atm class with its functionalities.

import datetime
class Atm:
    #class ke andar hum variable , ek special method __init__ ke andar karte hai
    # __init__ ek constructor hai =>  jiske andar ka code automatically execute hota hai, jab hum us class ka object banate hai.

    # __init__ python me ek already predefined magic method hai (python me isi ko constructor bolte hai).
    # magic method ko object call nhi karta hai. (ye automatically call hota hai.)
    

    def __init__(self):

        #self.pin="" (public) -> acess to all
        self.__pin=""  #(private) -> To hide data outside the class
        self.__balance=100 #(instance variable)
        self.__transactions = [] 

        self.menu() #(call menu)
    
    def get_pin(self):
       return self.__pin
    
    def set_pin(self,new_pin):
        self.__pin=new_pin
        print('pin changed')


    def menu(self):
        while True:
            user_input=input("""Hello, how would you like to proceed?
                                1.Enter 1 to create pin
                                2.Enter 2 to deposit
                                3.Enter 3 to withdraw
                                4.Enter 4 to check balance
                                5.Enter 5 to show transaction history
                                6.Enter 6 to change Your Atm pin number
                                7.Enter 7 to exit
                             
                             """)
            
            if user_input=='1':
               self.create_pin()
            elif user_input=='2':
                self.deposit()
            elif user_input=='3':
               self.withdraw()
            elif user_input=='4':
                self.check_balance()
            elif user_input == '5':
                self.show_transactions()
            elif user_input =='6':
                self.change_pin()
            elif user_input == '7':
                print("Bye...")
                break
            
    def create_pin(self):
        self.__pin=input('Enter Your Pin: ')
        print('Pin Set Successfully')


    def deposit(self):
      
      if self.__pin == "":
        print("Please create PIN first")
        return
      
      else:

        temp=input('Enter Your Pin: ')

        if temp==self.__pin:
            amount=int(input('Enter your amount'))
            self.__balance=self.__balance+amount
            self.__transactions.append(
                  f"{datetime.datetime.now()} | Deposited ₹{amount} | Balance ₹{self.__balance}"
            )
            print('Deposit Successfully')
        else:
            print('Invalid pin')
    

    def withdraw(self):
      if self.__pin == "":
        print("Please create PIN first")
        return
      else:
        temp=input('Enter Your Pin: ')
        if temp==self.__pin:
            amount=int(input('Enter your amount'))
            if amount < self.__balance:
                self.__balance=self.__balance-amount
                self.__transactions.append(
                  f"{datetime.datetime.now()} | Withdrawn ₹{amount} | Balance ₹{self.__balance}"
                )
                print('Withdraw Successfully')
            else:
                print('Insufficient Funds')
        else:
            print('invalid Pin')
    
    def show_transactions(self):
        if self.__pin == "":
            print("Please create PIN first")
            return

        temp = input('Enter Your Pin: ')
        if temp == self.__pin:
            if len(self.__transactions) == 0:
                print("No transactions yet")
            else:
                print("\nTransaction History:")
                for t in self.__transactions:
                    print(t)
        else:
            print("Invalid pin")

    def change_pin(self):
        old_pin = input("Enter old PIN: ")
        if old_pin == self.__pin:
            new_pin = input("Enter new PIN: ")
            self.__pin = new_pin
            print("PIN changed successfully")
        else:
            print("Wrong PIN")

    def check_balance(self):
      if self.__pin == "":
        print("Please create PIN first")
        return
      else:

        temp=input('Enter Your Pin: ')
        if temp==self.__pin:
            print('Your Current Balance is: ',self.__balance)
        else:
            print('Invalid pin')

sbi = Atm() #(yaha par, sbi ek reference variable hai, jisne object ko store kiya hai.)
#print(sbi.get_pin())

#jis object ke saath hum currently kaam kar rahe hai wahi 'self' hai.(print(id(self)))

# (double undersore) use karne se python me data hide ho jata  hai.
#Nothing in python is truly private.
```

```python
#Create a fraction class for performing operations on fraction:(Khud ka data types)

class Fraction:

    def __init__(self,n,d):
        self.num=n
        self.den=d

    def __str__(self):
        return "{}/{}".format(self.num,self.den)
    
    def __add__(self, other):
        
        temp_num= (self.num* other.den) + (other.num *  self.den)
        temp_den= self.den * other.den

        return "{}/{}".format(temp_num,temp_den)
    
    def __sub__(self, other):
        
        temp_num= (self.num* other.den) - (other.num *  self.den)
        temp_den= self.den * other.den

        return "{}/{}".format(temp_num,temp_den)
    
    def __mul__(self, other):
        
        temp_num= (self.num* other.den) * (other.num *  self.den)
        temp_den= self.den * other.den

        return "{}/{}".format(temp_num,temp_den)
    
    def __truediv__(self, other):
        
        temp_num= (self.num* other.den) / (other.num *  self.den)
        temp_den= self.den * other.den

        return "{}/{}".format(temp_num,temp_den)
    
x=Fraction(7,5) # yaha par x ek reference variable hai
y=Fraction(4,3)
print(x)
print(y)
print(x+y)
print(x-y)
print(x*y)
print(x/y)
print(type(x))

#jis variable me object store hota hai wo ek reference variable hota hai.

#In python, class ka objects, mutable hota hai (like list,dict,sets).
```

```python
#Creating a customer class (Collection of Object)


class Customer:
    counter=1#static/class Variable
    def __init__(self,name,age):
        self.name=name
        self.age=age

    #@staticmethod #yah ek static method hai jo class ke andar sabko access hoga.
    def intro(self):
        print('I am', self.name, " and I am", self.age)

    
c1=Customer('Raj',20)
c2=Customer('Abhinav',21)
c3=Customer('Ankita',19)

L=[c1,c2,c3]

print(Customer.counter)
print(c1.counter)
for i in L:
    i.intro()

#(1)instance variable:- yah ek aisa variable hota hai, jiska value har object ke liye different hota hai
# instance variable always constructor ke andar present hota hai (yah self se access hota hai)

#(2)static/class Variable:- yah constructor ke bahar, and class ke andar hota hai
# static variable ko {className. variableName} se access hota hai.
```

```python
#Aggregation:- Jab ek class dusri class ko use karti hai, lekin uska ownership strong nahi hota
#Dono objects independent hote hain
#Ek object delete ho jaye, dusra still exist kar sakta hai

class Teacher:
    def __init__(self, name, subject):
        self.name = name
        self.subject = subject

    def show(self):
        print(f"{self.name} teaches {self.subject}")


class Department:
    def __init__(self, dept_name, teachers):
        self.dept_name = dept_name
        self.teachers = teachers   # Aggregation

    def show_teachers(self):
        print(f"\nDepartment: {self.dept_name}")
        for teacher in self.teachers:
            teacher.show()


# Independent Teacher objects
t1 = Teacher("Raj", "Math")
t2 = Teacher("Ankita", "Physics")
t3 = Teacher("Abhinav", "Chemistry")

# Department uses existing teachers
dept = Department("Science", [t1, t2, t3])

dept.show_teachers()
```

```python
#Inheritance:-
#Jab ek class dusri class ke properties aur methods ko inherit (use) karti hai

#parent class ke private member ko access nhi kar sakte hai.
class Employee:
    def __init__(self, name):
        self.name = name

    def show(self):
        print("Name:", self.name)


class Manager(Employee):
    def __init__(self, name, department):
        super().__init__(name)   # parent constructor call
        self.department = department

    def show_details(self):
        print(self.name, "manages", self.department)


m = Manager("Raj", "IT")
m.show()           # inherited
m.show_details()   # ownSame method/function name
```

```python
#polymorphism:- Same method/function name;; Different behavior depending on object

class Animal:
    def sound(self):
        print("Animal makes sound")


class Dog(Animal):
    def sound(self):   # override
        print("Dog barks")


class Cat(Animal):
    def sound(self):   # override
        print("Cat meows")


a = Animal()
d = Dog()
c = Cat()

a.sound()
d.sound()
c.sound()

```
