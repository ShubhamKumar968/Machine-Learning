```python
#function:- It is a block of code which runs only when it is called

#function syntax:-  'def' keyword + function_name (parameter):
#Everything in python is a object {function are also an object}
def is_even(i):
  if type(i)==int:
    if i%2==0:
       return True
    return "odd"
  else:
    return 'Not allowed'
    
print(is_even('hello'))
is_even(6) #function call

for i in range(1,11):
    print(is_even(i))
```

```python
#Types of argument:-

def power(a,b):
    return a**b

print(power(2,3))

#(1)Default argument:-
def power1(a=1,b=1):
    return a**b

print(power1(3))
print(power1())
print(power1(3,4))

#(2)positional argument:-
print(power1(2,4))

#(3)keyword argument:-
print(power1(b=5,a=2))

#Note:- Priority of: (keyword argument > positional argument)

#(4)arbitrary argument:-
def flexible(*number):
    product=1
    print(number, type(number))
    for i in number:
        product=product*i
    
    print(product)

print(flexible(1,2,3,4,5))
```

```python
#some analysis on function using arguments:-

#(1)
def f1(a,b):
    print(a+b)

print(f1(3,5,7))
print(f1(3,5))
```

```python
#(2)
def f2(a,b,c):
    print(a+b+c)

print(f2(3,5,7))
print(f2(3,5))
```

```python
#(3)
def f3(a,b,c=0):
    print(a+b+c)

print(f3(3,5,7))
print(f3(3,5))
```

```python
#(4)
def f4(a,b=0,c):
    print(a+b+c)

print(f4(3,5,7))
print(f4(3,5))
```

```python
#(4)
def f5(a,b=0,c=0):
    print(a+b+c)

print(10)
print(f5(3,5,7))
print(f5(3,5))
```

```python
#(6)
def f6(a=0,b=0,c=0):
    print(a+b+c)

print(f6())
print(f6(3,5,7))
print(f6(3,5))
print(f6(3))
```

```python
#(7) 
def f7(a,b):
    print('a=',a,'b=',b)

print(f7(2,3)) #positional argument

print(f7(b=2,a=3))#keyword argument

print(f7(2,a=3))
```

```python
#(8)
def f8(a,b):
    print('a=',a,'b=',b)

print(f8(2,b=3))

print(f8(a=2,3))
```
