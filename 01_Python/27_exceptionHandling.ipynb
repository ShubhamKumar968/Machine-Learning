```python
#catching specific exception
try:
    f=open('sample.text','r')
    print(f.read())
   # print(m)
   # print(5/0)
    L=[1,2,3]
    print(L[5])
except FileNotFoundError:
    print('File Not found')
except NameError:
    print('Variable not defined')
except ZeroDivisionError:
    print("Can't divide by zero" )
except Exception as e: #generic exception is written in the last to handle any type of error.
    print(e)
```

```python
#else block:-
try:
     f=open('sample.text','r')
except FileNotFoundError:
     print('File not found')
except Exception:
     print('Kuchh to gadbad hai')
else:#agar try sahi se execute hoga to else run hoga, lekin agar try sahi se execute nhi hoga to koi except trigger hoga.
     print(f.read())
```

```python
#finally block (yah kisi bhi haal me execute hoga hi chahe kuchh bhi ho jaye)
try:
     f=open('sample.text','r')
except FileNotFoundError:
     print('File not found')
except Exception:
     print('Kuchh to gadbad hai')
else:#agar try sahi se execute hoga to else run hoga, lekin agar try sahi se execute nhi hoga to koi except trigger hoga.
     print(f.read())
finally:
     print('Ye to print hoga hi!!')

```

```python
#In java: {except->catch, raise->throw}

class Bank:

    def __init__(self,balance):

        self.balance=balance
    
    def withdraw(self,amount):
        if amount <0:
            raise Exception("amount can't be -ve")
        if self.balance <amount:
            raise Exception('Paise nhi hai tere paas')
        self.balance=self.balance-amount

obj=Bank(1000)

try:
    obj.withdraw(1500)

except Exception as e:
    print(e)
else:
    print(obj.balance)
```

```python
#creating custom exceptions:-
class MyException(Exception): #to call raise we have to inherited from Exception class
    def __init__(self,message):#(constructor)
        
        print(message)

class Bank:

    def __init__(self,balance):

        self.balance=balance
    
    def withdraw(self,amount):
        if amount <0:
            raise MyException("amount can't be -ve")
        if self.balance <amount:
            raise MyException('Paise nhi hai tere paas')
        self.balance=self.balance-amount

obj=Bank(1000)

try:
    obj.withdraw(1500)

except MyException as e:
    pass
else:
    print(obj.balance)
```

```python
#Simple example:- (for creating custom exceptions)

class SecurityError(Exception):

  def __init__(self,message):
    print(message)

  def logout(self):
    print('logout')

class Google:

  def __init__(self,name,email,password,device):
    self.name = name
    self.email = email
    self.password = password
    self.device = device

  def login(self,email,password,device):
    if device != self.device:
      raise SecurityError('bhai teri to lag gayi')
    if email == self.email and password == self.password:
      print('welcome')
    else:
      print('login error')



obj = Google('nitish','nitish@gmail.com','1234','android')

try:
  obj.login('nitish@gmail.com','1234','windows')
except SecurityError as e:
  e.logout()
else:
  print(obj.name)
finally:
  print('database connection closed')
```
