```python
import random

#(1) randint(a,b) -> generate random integer in range of a and b (included)
x=random.randint(2,9)
print(x)

#(2) randrange(a,b) -> here in random integer b is not included
y=random.randrange(1,5)
print(y)

#(3) random() -> It generate random decimal till 16 place between 0 and 1
z=random.random()
print(z)

#(4) choice([1,2,3,4]) -> generate random from given list
a=random.choice([1,5,2,4,8,6])
print(a)
```
