```python
#dictionary:-  (1)It has no indexing
#(2)It is a mutable object
#(3){key=>immutable, value->can be mutable}
#(4)keys must be unique, value may be same

#mutable->list,sets,dictionary{memory me edit ke baad address same rahega}
#immutable->string, tuple, int, float, boolean, complex

d={1:'Rahul', 2:'sohan', 3:'Rohit'}
print(d)

d1={(1,2,3,):'Rahul'}
print(d1)
```

```python
#Accessing item in dictionary (access item using keys)
d={1:'Rahul', 2:'sohan', 3:'Rohit','age':16}
print(d[1])
print(d['age'])
print(d.get(2))

d1={1:'Rahul', 2:'sohan', 3:'Rohit','age':16,'marks':{'math':92,'Science':93,'DSA':99}}
print(d1['marks']['DSA'])
```

```python
#Add new key-value pairs
d={1:'Rahul', 2:'sohan', 3:'Rohit','age':16}
d['marks']=99

print(d)

#edit item
d[2]='Ram'
print(d)
```

```python
#delete item 
d={1:'Rahul', 2:'sohan', 3:'Rohit','age':16}

del d[3]
print(d)

d.pop(2)
print(d)

d.clear()
print(d)
```

```python
#dictionary operations

# (d1+d2) and (d*3) is not applicable

d={1:'Rahul', 2:'sohan', 3:'Rohit','age':16}
for i in d:
    print(i,'->',d[i])

print(1 in d)

```

```python
#functions in dictionary

d={1:'Rahul', 2:'sohan', 3:'Rohit','age':16}

print(d.keys())
print(d.values())

print(sorted(d, reverse=True))
#min(),max(),len() are also applicable
```
