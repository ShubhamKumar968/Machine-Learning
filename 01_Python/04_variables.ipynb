```python
#Variables :- It is containers used to hold data during program execution

#In python , we do not to define datatype of variable,
# it is decided at runtime , that what type of value are we assigning to it.

name ='Shubham'
print(name)

# So In python , Creation of variable is "Dynamic Typing".
```

```python
# In python, we can change data as well as type of a variable at runtime

name="Shubham"
print(name)
print(type(name))

#after modification => called as "Dynamic Binding"
name=3+4j
print(name)
print(type(name))
```
