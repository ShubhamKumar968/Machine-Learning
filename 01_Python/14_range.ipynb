```python
#Range:- 

#It is a class.(iterables)
#it contains only int type value
#it is immutable sequence(not changable)
#it contains sequence of integers with common difference (A.P)

# Syntax: range(beg, end, step)
# by default if not mentioned :- {beg=0,step=1}
#last element is not included

print(list(range(3))) #range(end)

print(list(range(1,10))) #range(beg,end)

print(list(range(2,10,3))) #range(beg,end,+ve step)

print(list(range(10,3,-2)))#range(beg,end,-ve step)

```
