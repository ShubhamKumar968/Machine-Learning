```python
#tuples:-it is immutable object, it can store heterogeneous items

t1=(1,2,3,"hello",)
print(t1)

t2=(1)
print(type(t2))

t3=(1,)
print(type(t3))

t4=(1,2,3,(4,5))
print(type(t4))

t5=tuple('goa')
print(t5)

print(tuple([1,2,3,4,5]))
```

```python
#Accessing tuple items

t=(1,3,4,5,'hii',True,(4,5))

print(t[0])
print(t[1])

print(t[6][1])

print(t[-1][0])
```

```python
#Edit and delete tuple elements

#tuple is an immutable object so it can't change or edited,deleted once created.
t=(1,2,3,4)

i=0
while i<len(t):
    print(t[i])
    i+=1

t[0]=100
print(t)
```

```python
#Operations on tuple:
t1=(1,2,3,)
t2=(4,5,6,)
t3=t1+t2
print(t3)

print(1 in t3)
print(t1*4)
print(t1>t2)
for i in t2:
    print(i)
```

```python
#functions in tuple
#jaha par indexing hota hai waha slicing operator use kar sakte hai

t=(1,2,3,5,0)

print(max(t))

print(len(t))

print(sum(t))

print(sorted(t))

print(sorted(t,reverse=True))

print(t.count(3))

print(t[: : -1])
#So, Tuples are read only data type (data integrity is maintained)
```

```python
#user input in tuple

t1=tuple( [int (e) for e in input('Enter the number').split(',')])
print(t1)
```
