```python
#Benefit: To load a big file in memory

big_L=['Hello ' for i in range(1000)]

with open('big.txt', 'w') as f:
    f.writelines(big_L)
```

```python
#working on data to process it chunk_wise
with open('big.txt', 'r') as f:

    chunk_size=10
    #yaha par hum 100 characerko chunk by chunk lakar memory me load kar raha hu jab tak khatam na ho
    while len(f.read(chunk_size)) > 0:
        print(f.read(chunk_size), end='***')
        f.read(chunk_size)#To load next chunk (to avoid it from infinite loop)
```

```python
#Serialization:- Process of converting python data types to JSON format(same as dictionary)

#JASON(JavaScript on Notation):- It is understood by every programming language(Universal) 

import json
#list
L=[1,2,3,4,'Hello',True]

with open('list_demo.json','w') as f:
    json.dump(L,f)
```

```python
#dictionary

d={
    'name' : 'Ram',
    'age' : 20,
    'Gender' : 'Male'
}

with open('dict_demo.json','w') as f:
    json.dump(d,f,indent=4)
```

```python
#Deserialization:-Process of converting JSON format to python data types 

import json
#vvery important for handle complex data type
with open('dict_demo.json','r') as f:
    d=json.load(f)
    print(d)
    print(type(d))
```

```python
#dumping custom object into file
class Person:

  def __init__(self,fname, lname, age,gender):
      
      self.fname=fname
      self.lname=lname
      self.age=age
      self.gender=gender

```

```python
person=Person('Mohan','Kumar', 33, 'male')
```

```python
#serialization
#As a dictionary

import json
def show_object(person):
    if isinstance(person,Person):
        return {'name':person.fname + '' + person.lname, 'Age':person.age, 'Gender' :person.gender }

with open('class_demo.json','w') as f:
    json.dump(person,f, default=show_object,indent=3)
```

```python
#deserializaton
import json

with open('class_demo.json','r') as f:
    d=json.load(f)
    print(d)
    print(type(d))
```

```python
#As a string
import json
def show_object(person):
    if isinstance(person,Person):
        return '{} {} age->{} gender -> {}'.format(person.fname,person.lname,person.age,person.gender)

with open('class_demo.json','w') as f:
    json.dump(person,f, default=show_object,indent=3)
```

```python
#Pickling:- It convert python object into Byte (Binary files)-> transferable
class Person:

  def __init__(self,name, age):
      
      self.name=name
      self.age=age

  def display_info(self):
      print('Hii my name is', self.name,'and I am ', self.age, 'Years old')
```

```python
p=Person('Nitesh',33)
```

```python
#picke dump

import pickle

with open('person.pkl','wb') as f:
    pickle.dump(p,f)
```

```python
#picke load

import pickle

with open('person.pkl','rb') as f:
    pickle.load(f)

print(p.display_info())

#Pickle store data in binary format,, while json store data in the text(human readable) format
```
